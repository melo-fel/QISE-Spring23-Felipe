1. Define a qubit and explain the difference between a classical bit and a qubit.

- Answer: A qubit is the basic unit of information in quantum computing. The difference between a qubit and a bit is that a classical bit can only be represented
    as a 1 or 0, as it is a fixed piece of data with a value, while a qubit has superposition when it comes to being either 0 or 1. Truly, a qubit is more correctly
    characterized as being either 0 or 1 or even in between, a qubit is allowed to be in these multiple states until it gets measured.

2. Explain the concept of superposition in quantum computing.

- Answer: The concept of superposition in quantum computing is the ability of being able to be in multiple states at once until you measure a system or quantum
    value. As such, it is more easily explained in the concept of Shrödinger's cat as the thought experiment centers on the result that a cat in a box filled
    with a radioactive substance can be both dead and alive until one opens that box to have an outcome be completed in reality. For qubits specifically, superpositon can
    also be explained with the double-slit experiment, where one can observe that a particle is able to encompass the abilities of a particle and a wave when observed, which
    also explains the importance of measurement in quantum computing.

3. Describe the process of measurement in quantum computing and its effect on a qubit.

- Answer: In the process of measurement, one will essentially create a quantum circuit for a qubit, enact certain gates upon that qubit, and then use measurement
    to figure out a numerical value to evaluate what the qubit's position between 0 and 1 will be. As such, measurement's effect on a qubit is that it essentially
    "breaks" the qubit since measurement is an interference to the quantum particle, and we can only measure the possibility of the particle being either 0 or 1
    to get a value that can be used to send data and information.

4. What is a quantum gate? Provide an example of a single qubit gate and explain its operation.

- Answer: A quantum gate is the quantum version of classical logic gates, wherein they are used to enact an effect on a qubit like a logic gate enacts a change
    on a bit. An example of a quantum gate would be the X gate, where it changes a qubit to the opposite value from what it was originally set. If the qubit was set in the 0 state,
    the X gate will make it turn to be in the 1 state and vice versa.

4. Using qiskit, create a quantum circuit that generates a Bell state. Run the circuit using the qasmsimulator backend, and print the measurements counts.

```python
from qiskit import Aer, execute
bell = QuantumCircuit(2,2)
bell.h(0)
bell.cnot(0, 1)
bell.measure(0, 0)
bell.measure(1, 1)

backend = Aer.get_backend('qasm_simulator')
job = backend.run(transpile(bell, backend), shots=1024)
result = job.result()
counts = result.get_counts(bell)
print(counts)
```

5. Plot a Bloch sphere representation of the state 1√2 |0⟩ + 1/√2 |1⟩ using qiskit.

```python
fourth = QuantumCircuit(1, 1)
fourth.h(0)
fourth.measure(0, 0)

from qiskit.tools.visualization import plot_bloch_multivector, plot_bloch_vector
from qiskit import Aer, execute

sim_statevector = Aer.get_backend('aer_simulator_statevector')
job_statevector = sim_statevector.run(fourth, shots=1)

backend = Aer.get_backend('statevector_simulator')
job = execute(fourth, backend=backend, shots=1, memory=True)
job_result = job.result()


plot_bloch_multivector(job_result.get_statevector(fourth))
```
